```python
df['mpg'] = df['mpg'].astype(int)
df['displacement'] = df['displacement'].astype(int)
df['acceleration'] = df['acceleration'].astype(int)
```

```python
X = df[['cylinders', 'displacement', 'horsepower', 'weight',
       'acceleration', 'model year', 'origin']]
y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

std = StandardScaler()
X_std = std.fit_transform(X)

x_train,x_test,y_train,y_test = train_test_split(X_std,y,test_size = 0.2,random_state = 42)

model = LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

print('mse',mean_squared_error(y_test,y_pred))
print('mae',mean_absolute_error(y_test,y_pred))
print('r2_score',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

poly = PolynomialFeatures()
X_poly = poly.fit_transform(X_std)

x_train,x_test,y_train,y_test = train_test_split(X_poly,y,test_size = 0.2,random_state = 42)

model = LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

print('mse',mean_squared_error(y_test,y_pred))
print('mae',mean_absolute_error(y_test,y_pred))
print('r2',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Lasso,Ridge
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV

lasso = Lasso()
params = {
    'alpha' : [0.001,0.01,0.1,1,10,100],
    'max_iter' : [10000]
}

grid = GridSearchCV(lasso,params,cv = 5,scoring = 'r2')
grid.fit(x_train,y_train)

y_pred_lasso = grid.predict(x_test)
print('mse',mean_squared_error(y_test,y_pred_lasso))
print('mae',mean_absolute_error(y_test,y_pred_lasso))
print('r2',r2_score(y_test,y_pred_lasso))


# assume: poly is your fitted PolynomialFeatures instance
# and grid is your fitted GridSearchCV

best_lasso = grid.best_estimator_

# get the coefs (if grid was fitted directly on transformed X)
coefs = best_lasso.coef_            # shape (n_transformed_features,)

# get transformed feature names
feature_names = poly.get_feature_names_out(X.columns)  # X is the original DataFrame with 7 cols

import pandas as pd
coef_series = pd.Series(coefs, index=feature_names)

selected_transformed = coef_series[coef_series != 0]
print("Selected transformed features (non-zero coefficients):")
print(selected_transformed)
```

```python
ridge = Ridge()

random = RandomizedSearchCV(ridge,params,cv = 5,scoring = 'r2',n_iter = 4)
random.fit(x_train,y_train)

y_pred = random.predict(x_test)

print('mse',mean_squared_error(y_test,y_pred))
print('mae',mean_absolute_error(y_test,y_pred))
print('r2',r2_score(y_test,y_pred))
```

```python
# Lasso is the best

import joblib

# Save the specific objects you fitted on your training data
joblib.dump(std, 'std_scaler.pkl')   # The StandardScaler used on x_train
joblib.dump(poly, 'poly_feats.pkl')     # The PolynomialFeatures used on x_train
joblib.dump(grid.best_estimator_, 'lasso_model.pkl') # The trained model

print("All components saved.")
```
